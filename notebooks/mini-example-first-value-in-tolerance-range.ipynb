## Minimal example for finding first value in a given tolerance range

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt

# create some dummy time series
x = np.linspace(0,1,100)
y = -np.sin(x*50)
y *= np.sin(x*10)**2

# plot the raw signal
plt.figure(figsize=(10,5))
plt.plot(y, marker='.', ls=':')
plt.grid()

# find values that are in a certain range around 0.5 and mask others
# one need to choose the tolerance range big enough such that enough data points lie within that range.
tolerance = (0.42, 0.56)
y_ = np.where(np.logical_and(y > tolerance[0], y < tolerance[1]), y, np.nan)
plt.gca().axhspan(*tolerance,alpha=.2)
# set all these values to common value
y_[~np.isnan(y_)] = 1

# now find the first occurrence in that array which is not NAN
ym = np.nanargmin(y_)
plt.gca().plot(ym, y[ym], marker='P', ms=10, ls='')
```
